Procesamiento Temperatura

```python
df_TP = pd.read_csv('estaciones metereologicas con empresa distribuidora de energia.csv', dtype=str)
df_TP['anio'] = pd.to_numeric(df_TP['anio'], errors='coerce')

# Elimina filas donde 'anio' es NaN después de la conversión (si las hubiera)
df_TP.dropna(subset=['anio'], inplace=True)
df_filtrado = df_TP[(df_TP['anio'] >= 2000) & (df_TP['anio'] <= 2023)]
```

```python
df_filtrado
```

```python
id_vars = [col for col in df_filtrado.columns if col not in ['ene', 'feb', 'mar', 'abr', 'may', 'jun', 
                                                    'jul', 'ago', 'sep', 'oct', 'nov', 'dic']]

# Convertir de formato ancho a largo
df_long = pd.melt(df_filtrado, 
                  id_vars=id_vars, 
                  value_vars=['ene', 'feb', 'mar', 'abr', 'may', 'jun', 
                             'jul', 'ago', 'sep', 'oct', 'nov', 'dic'],
                  var_name='mes_nombre', 
                  value_name='temperatura')

# Crear un diccionario de mapeo para los códigos de mes
mes_codigo = {
    'ene': 1, 'feb': 2, 'mar': 3, 'abr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'ago': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dic': 12
}
# Agregar la columna de código de mes
df_long['mes_codigo'] = df_long['mes_nombre'].map(mes_codigo)

# Reorganizar las columnas si es necesario
# Poner mes_nombre y mes_codigo al principio después de las columnas identificadoras principales

                
df_long['temperatura'] = pd.to_numeric(df_long['temperatura'], errors='coerce')
```

```python
df_promedio = df_long.groupby(['anio', 'mes_nombre', 'mes_codigo','IDSISDAT'])['temperatura'].mean().round(2).reset_index()
df_promedio.to_csv('promedio_temperatura_mensual.csv', index=False)
```

Interpolacion del PIB trimestral

```python
df = pd.read_excel('4.5.2-Serie_histórica-2023 (1).xlsx')
df
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
from scipy.interpolate import interp1d

# Función para procesar el dataset del PIB trimestral y generar interpolación mensual
def procesar_pib_trimestral(df):
    
    # 1. Preparar el dataset original
    df_original = df.copy()

    
    # Convertir la columna PIB a numérico (remover comas si existen)
    if df_original['Pib_real'].dtype == 'object':
        df_original['Pib_real'] = df_original['Pib_real'].str.replace(',', '').astype(float)
    
    # 2. Procesar las fechas trimestrales y convertir a formato datetime
    def convertir_fecha_trimestral(fecha_str):
        """Convierte formato '2000.I' a datetime del primer mes del trimestre"""
        año, trimestre = fecha_str.split('.')
        año = int(año)
        
        # Mapear trimestre a mes
        trimestre_a_mes = {'I': 1, 'II': 4, 'III': 7, 'IV': 10}
        mes = trimestre_a_mes[trimestre]
        
        return datetime(año, mes, 1)
    
    # Aplicar conversión de fechas
    df_original['fecha_datetime'] = df_original['Fecha'].apply(convertir_fecha_trimestral)
    
    # 3. Crear dataset mensual completo (2000-2023)
    fecha_inicio = datetime(2000, 1, 1)
    fecha_fin = datetime(2023, 12, 1)
    
    # Generar todas las fechas mensuales
    fechas_mensuales = pd.date_range(start=fecha_inicio, end=fecha_fin, freq='MS')
    
    # Crear DataFrame mensual base
    df_mensual = pd.DataFrame({
        'fecha': fechas_mensuales
    })
    
    # Agregar columnas de año, mes_nombre y mes_id
    df_mensual['anio'] = df_mensual['fecha'].dt.year
    df_mensual['mes_nombre'] = df_mensual['fecha'].dt.month_name()
    df_mensual['mes_id'] = df_mensual['fecha'].dt.month  # Enero=1, Febrero=2, ..., Diciembre=12
    
    # 4. Realizar interpolación
    # Preparar datos para interpolación
    fechas_trimestrales_num = [d.timestamp() for d in df_original['fecha_datetime']]
    pib_trimestral = df_original['Pib_real'].values
    
    f_interpolacion = interp1d(fechas_trimestrales_num, pib_trimestral, 
                              kind='cubic', bounds_error=False, fill_value='extrapolate')    
    fechas_mensuales_num = [d.timestamp() for d in df_mensual['fecha']]
    pib_mensual_interpolado = f_interpolacion(fechas_mensuales_num)
    
    df_mensual['PIB_mensual_interpolado'] = pib_mensual_interpolado
    
    # 5. Agregar datos trimestrales originales donde corresponda
    df_mensual['PIB_trimestral_original'] = np.nan
    df_mensual['es_dato_trimestral'] = False
    
    # Marcar y llenar datos trimestrales originales
    for _, row in df_original.iterrows():
        fecha_trimestral = row['fecha_datetime']
        mask = df_mensual['fecha'] == fecha_trimestral
        if mask.any():
            df_mensual.loc[mask, 'PIB_trimestral_original'] = row['Pib_real']
            df_mensual.loc[mask, 'es_dato_trimestral'] = True
    
    # 6. Formatear el DataFrame final
    df_final = df_mensual.copy()
    df_final['fecha_str'] = df_final['fecha'].dt.strftime('%Y-%m')
    
    # Reordenar columnas
    columnas_finales = [
        'fecha_str', 'fecha', 'anio', 'mes_nombre', 'mes_id',
        'PIB_mensual_interpolado', 'PIB_trimestral_original', 'es_dato_trimestral'
    ]
    df_final = df_final[columnas_finales]
    
    # Redondear valores PIB
    df_final['PIB_mensual_interpolado'] = df_final['PIB_mensual_interpolado'].round(2)
    
    return df_final

# Función para mostrar estadísticas del resultado
def mostrar_estadisticas(df_resultado):
    """Muestra estadísticas del dataset procesado"""
    print("=== ESTADÍSTICAS DEL DATASET PROCESADO ===")
    print(f"Total de registros mensuales: {len(df_resultado)}")
    print(f"Período: {df_resultado['fecha'].min().strftime('%Y-%m')} a {df_resultado['fecha'].max().strftime('%Y-%m')}")
    print(f"Datos trimestrales originales: {df_resultado['es_dato_trimestral'].sum()}")
    print(f"Datos interpolados: {(~df_resultado['es_dato_trimestral']).sum()}")
    print(f"PIB mínimo: {df_resultado['PIB_mensual_interpolado'].min():,.0f}")
    print(f"PIB máximo: {df_resultado['PIB_mensual_interpolado'].max():,.0f}")
    print("\n=== PRIMEROS 10 REGISTROS ===")
    print(df_resultado.head(10).to_string(index=False))
    print("\n=== ÚLTIMOS 10 REGISTROS ===")
    print(df_resultado.tail(10).to_string(index=False))

# Ejemplo de uso:
# df_resultado = procesar_pib_trimestral(df)  # df es tu DataFrame original
# mostrar_estadisticas(df_resultado)

# Para ver solo datos trimestrales originales:
# datos_trimestrales = df_resultado[df_resultado['es_dato_trimestral'] == True]
# print(datos_trimestrales)

# Para ver solo datos interpolados:
# datos_interpolados = df_resultado[df_resultado['es_dato_trimestral'] == False]
# print(datos_interpolados)
```

```python
import pandas as pd
import glob
import os
import re

# ---------------------------------------------------------------------------
# 1. CONFIGURACIÓN: ¡AJUSTA ESTAS VARIABLES SEGÚN TUS NECESIDADES!
# ---------------------------------------------------------------------------
# El script asumirá que los archivos .xls están en el mismo directorio que este script.
# PATRON_DE_ARCHIVOS: Patrón para encontrar tus archivos Excel.
PATRON_DE_ARCHIVOS = '*.xlsx'  # Para todos los archivos .xls en el directorio actual
# PATRON_DE_ARCHIVOS = 'Reporte_Anual_*.xls' # Ejemplo para nombres más específicos

# NOMBRE_COLUMNA_ANIO_EN_XLS: Nombre de la columna que contiene el año DENTRO de los archivos Excel.
NOMBRE_COLUMNA_ANIO_EN_XLS = 'anio'  # ¡AJUSTA ESTO si el nombre es diferente!

# EXTRAER_ANIO_DEL_NOMBRE_ARCHIVO:
# True: si la columna de año NO está en los Excel y se debe obtener del nombre del archivo.
# False: si la columna de año YA EXISTE dentro de cada Excel.
EXTRAER_ANIO_DEL_NOMBRE_ARCHIVO = False  # ¡AJUSTA ESTO!

# NOMBRE_HOJA_EXCEL: Nombre o índice de la hoja de Excel a leer de cada archivo.
# - 0 para la primera hoja.
# - 'NombreDeLaHoja' si conoces el nombre.
# - None para leer todas las hojas (devolvería un diccionario de DataFrames por archivo,
#   lo que requeriría un manejo adicional para combinarlos si esa es la intención).
#   Para una unión simple de tablas, es mejor que todas estén en la misma hoja con el mismo nombre/índice.
NOMBRE_HOJA_EXCEL = 0  # Lee la primera hoja por defecto. ¡AJUSTA SI ES NECESARIO!

# NOMBRE_ARCHIVO_SALIDA: Nombre para el archivo combinado.
# Puedes guardarlo como .xlsx o .csv
NOMBRE_ARCHIVO_SALIDA_EXCEL = 'datos_historicos_combinados.xlsx'
# NOMBRE_ARCHIVO_SALIDA_CSV = 'datos_historicos_combinados.csv'

# ---------------------------------------------------------------------------

# Obtener la lista de archivos Excel en el directorio actual que coincidan con el patrón
# os.getcwd() obtiene el directorio de trabajo actual.
directorio_actual = os.getcwd()
ruta_busqueda = os.path.join(directorio_actual, PATRON_DE_ARCHIVOS)
todos_los_archivos = glob.glob(ruta_busqueda)

if not todos_los_archivos:
    print(f"No se encontraron archivos Excel ('{PATRON_DE_ARCHIVOS}') en el directorio: '{directorio_actual}'.")
    print("Por favor, verifica el patrón y asegúrate de que los archivos estén presentes.")
else:
    print(f"Archivos Excel encontrados ({len(todos_los_archivos)}):")
    for f_path in todos_los_archivos[:5]:  # Muestra los primeros 5 para verificar
        print(f" - {os.path.basename(f_path)}")
    if len(todos_los_archivos) > 5:
        print(f" - ... y {len(todos_los_archivos) - 5} más.")

    lista_de_dataframes = []

    for ruta_archivo in todos_los_archivos:
        nombre_archivo = os.path.basename(ruta_archivo)
        print(f"\nProcesando archivo: {nombre_archivo}...")
        try:
            # Leer el archivo Excel. Especifica la hoja si es necesario.
            # El motor 'xlrd' se usará automáticamente para .xls
            df_temporal = pd.read_excel(ruta_archivo, sheet_name=NOMBRE_HOJA_EXCEL)

            # --- MANEJO DE LA INFORMACIÓN DEL AÑO ---
            if EXTRAER_ANIO_DEL_NOMBRE_ARCHIVO:
                # Intenta encontrar 4 dígitos consecutivos (comúnmente un año)
                match = re.search(r'(\d{4})', nombre_archivo)
                if match:
                    anio_extraido = int(match.group(1))
                    # Añade la columna de año (o el nombre que hayas configurado)
                    df_temporal[NOMBRE_COLUMNA_ANIO_EN_XLS] = anio_extraido
                    print(f"  Año {anio_extraido} extraído del nombre de archivo y añadido como columna '{NOMBRE_COLUMNA_ANIO_EN_XLS}'.")
                else:
                    print(f"  ADVERTENCIA: No se pudo extraer el año del nombre de archivo: {nombre_archivo}. Este archivo podría no tener la columna de año.")
            else:
                # El año ya debería estar como columna en el Excel
                if NOMBRE_COLUMNA_ANIO_EN_XLS not in df_temporal.columns:
                    print(f"  ADVERTENCIA: La columna de año '{NOMBRE_COLUMNA_ANIO_EN_XLS}' no se encontró en {nombre_archivo} (Hoja: {NOMBRE_HOJA_EXCEL}).")
                    print(f"  Columnas encontradas: {df_temporal.columns.tolist()}")
                else:
                    print(f"  La columna de año '{NOMBRE_COLUMNA_ANIO_EN_XLS}' ya existe en el archivo.")
                    # Opcional: convertir a tipo entero si es necesario
                    try:
                        df_temporal[NOMBRE_COLUMNA_ANIO_EN_XLS] = pd.to_numeric(df_temporal[NOMBRE_COLUMNA_ANIO_EN_XLS], errors='coerce').astype('Int64')
                    except Exception as e_astype:
                        print(f"    No se pudo convertir la columna '{NOMBRE_COLUMNA_ANIO_EN_XLS}' a entero: {e_astype}")
            
            lista_de_dataframes.append(df_temporal)
        except FileNotFoundError:
            print(f"  ERROR: El archivo {nombre_archivo} no fue encontrado en la ruta especificada.")
        except ImportError as e_import:
             print(f"  ERROR: Necesitas instalar una biblioteca para leer archivos Excel. Para .xls, instala 'xlrd': pip install xlrd. Detalle: {e_import}")
             break # Detener si falta una dependencia crucial
        except Exception as e:
            print(f"  ERROR al leer o procesar el archivo {nombre_archivo} (Hoja: {NOMBRE_HOJA_EXCEL}): {e}")

    if lista_de_dataframes:
        print("\nConcatenando todos los DataFrames...")
        df_combinado = pd.concat(lista_de_dataframes, ignore_index=True, join='outer')

        print("\n--- Primeras filas del DataFrame combinado ---")
        print(df_combinado.head())
        print("\n--- Información del DataFrame combinado ---")
        df_combinado.info()
        print(f"\nDimensiones del DataFrame combinado (filas, columnas): {df_combinado.shape}")

        # Opcional: Verificar duplicados
        num_duplicados = df_combinado.duplicated().sum()
        if num_duplicados > 0:
            print(f"\nADVERTENCIA: Se encontraron {num_duplicados} filas duplicadas exactas.")
            # df_combinado.drop_duplicates(inplace=True, ignore_index=True) # Descomentar para eliminar
        else:
            print("\nNo se encontraron filas duplicadas exactas.")

        # --- GUARDAR EL DATAFRAME COMBINADO ---
        try:
            # Guardar como archivo Excel (.xlsx, que es más moderno)
            df_combinado.to_excel(os.path.join(directorio_actual, NOMBRE_ARCHIVO_SALIDA_EXCEL), index=False)
            print(f"\n¡Éxito! DataFrame combinado guardado como: '{os.path.join(directorio_actual, NOMBRE_ARCHIVO_SALIDA_EXCEL)}'")

            # Opcionalmente, guardar también o en su lugar como CSV:
            # df_combinado.to_csv(os.path.join(directorio_actual, NOMBRE_ARCHIVO_SALIDA_CSV), index=False)
            # print(f"\n¡Éxito! DataFrame combinado guardado como: '{os.path.join(directorio_actual, NOMBRE_ARCHIVO_SALIDA_CSV)}'")
        except Exception as e:
            print(f"\nERROR al guardar el archivo combinado: {e}")
    else:
        print("\nNo se procesaron DataFrames, el DataFrame combinado está vacío o no se pudo crear.")
```
